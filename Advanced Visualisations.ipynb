```python
#KDE Plot
k1 = sns.lmplot(data=movies, x= 'CriticRating', y='AudienceRating', fit_reg=False, hue="Genre", aspect=2 )
plt.show()
```

```python
sns.set_style("dark")
k1 = sns.kdeplot(data=movies, x="BudgetMillions", y = "CriticRating", shade=True)
plt.show()
```

```python
k2 = sns.kdeplot(data=movies, x="BudgetMillions", y = "AudienceRating", color="Red")
plt.show()
```

```python
#subplot
f, axes = plt.subplots(1,2, figsize=(12,6), sharex=True)
k1 = sns.kdeplot(data=movies, x="BudgetMillions", y = "CriticRating", shade=True, ax=axes[0])
k2 = sns.kdeplot(data=movies, x="BudgetMillions", y = "AudienceRating", color="Red", ax=axes[1])
k1.set(xlim=(-20,200))
plt.show()
```
